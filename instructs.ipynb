```python
def get_title(article_title:str):
    if 'Обзор' in article_title:
        a = list(filter(lambda x: x!='',article_title.split(': Обзор')))
        if a[-1]==' мобильной версии' or a[-1]==' (мобильная версия)':
            a[0]+=' мобильная версия'
            a=a[:1]
        if a[-1]==' первого эпизода':
            a[0]+=' первый эпизод'
            a=a[:1]
        if a[-1]==' второго эпизода':
            a[0]+=' второй эпизод'
            a=a[:1]
        if a[-1]==' кампании':
            a[0]+=' кампания'
            a=a[:1]
        if a[-1]==' мультиплеера':
            a[0]+=' мультиплеер'
            a=a[:1]
        if a[-1]==' (PC версия)' or a[-1]==' PC версии':
            a[0]+=' PC версия'
            a=a[:1]
        if a[-1]==' PvP-сервиса Demonion':
            a[0]+=' PvP-сервис Demonion'
            a=a[:1]
        if a[-1]==' дополнения Horizons':
            a[0]+=' дополнение Horizons'
            a=a[:1]
        if a[-1]==' всего сезона':
            a[0]+=' весь сезон'
            a=a[:1]
        if len(a)>1:
            print(article_title.split(': Обзор'))
        return a[0]
    if 'Видеообзор' in article_title:
        a = list(filter(lambda x: x!='',article_title.split(': Видеообзор')))
        if a[-1]==' мобильной версии' or a[-1]==' (мобильная версия)':
            a[0]+=' мобильная версия'
            a=a[:1]
        if a[-1]==' первого эпизода':
            a[0]+=' первый эпизод'
            a=a[:1]
        if a[-1]==' второго эпизода':
            a[0]+=' второй эпизод'
            a=a[:1]
        if a[-1]==' кампании':
            a[0]+=' кампания'
            a=a[:1]
        if a[-1]==' мультиплеера':
            a[0]+=' мультиплеер'
            a=a[:1]
        if a[-1]==' (PC версия)' or a[-1]==' PC версии':
            a[0]+=' PC версия'
            a=a[:1]
        if a[-1]==' PvP-сервиса Demonion':
            a[0]+=' PvP-сервис Demonion'
            a=a[:1]
        if a[-1]==' дополнения Horizons':
            a[0]+=' дополнение Horizons'
            a=a[:1]
        if a[-1]==' всего сезона':
            a[0]+=' весь сезон'
            a=a[:1]
        if len(a)>1:
            print(article_title.split(': Видеообзор'))
        return a[0]
    return
```

```python
data['title'] = data['title'].apply(lambda x:get_title(x)).sort_values(ascending=False)
```

```python
prefixes=['Расскажи о ','Что ты знаешь о ','Обзор ','']
```

```python
result=pd.DataFrame(columns=['instruct','answer'])
```

```python
for prefix in prefixes:
    temp = data.copy()
    temp['title']=prefix+temp['title']
    temp.rename(columns = {'title':'instruct','review':'answer'},inplace=True)
    result = pd.concat([result,temp])
```
